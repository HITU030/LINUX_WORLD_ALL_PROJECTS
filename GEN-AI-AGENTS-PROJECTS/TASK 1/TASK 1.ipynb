# TASK 1  : AI Resume & Cover Letter Generator

```python
import ollama
import gradio as gr

MODEL = "llama3.2:latest"

def ollama_call(prompt):
    return ollama.chat(
        model=MODEL,
        messages=[{"role": "user", "content": prompt}]
    )["message"]["content"]


def generate(name, phone, email, linkedin, github, location,
             role, education, programming, libraries,
             tools, databases, platforms,
             project1, project2, project3,
             internship, company_name, hiring_role):

    resume_prompt = f"""
Create a professional ATS-friendly resume.

Use ONLY the provided information.
Do not add fake skills, projects, internships or achievements.

Name: {name}
Phone: {phone}
Email: {email}
LinkedIn: {linkedin}
GitHub: {github}
Location: {location}

Role: {role}
Education: {education}

Programming: {programming}
Libraries: {libraries}
Tools: {tools}
Databases: {databases}
Platforms: {platforms}

Projects:
1. {project1}
2. {project2}
3. {project3}

Internship:
{internship}

Generate:
- Header
- Career Objective
- Skills
- Projects (2-3 lines each + Technologies)
- Internship (2-3 lines)
- Education
"""



    
    cover_prompt = f"""
Write a professional cover letter.
Candidate:
{name}

Phone: {phone}
Email: {email}
Location: {location}

Applying For:
{hiring_role}

Company:
{company_name}

Education:
{education}

Programming:
{programming}

Libraries:
{libraries}

Tools:
{tools}

Databases:
{databases}

Platforms:
{platforms}

Projects:
{project1}
{project2}
{project3}

Internship:
{internship}

Professional tone.
Mention skills, projects and internship briefly.
End professionally.
"""

    
    resume = ollama_call(resume_prompt)
    cover = ollama_call(cover_prompt)
    open("Resume.txt","w",encoding="utf-8").write(resume)
    open("CoverLetter.txt","w",encoding="utf-8").write(cover)
    return resume, cover, "Resume.txt", "CoverLetter.txt"


with gr.Blocks(title="AI Resume & Cover Letter Generator") as resume_builder:

    gr.Markdown("# AI Resume & Cover Letter Generator")

    name = gr.Textbox(label="Full Name")
    phone = gr.Textbox(label="Phone")
    email = gr.Textbox(label="Email")
    location = gr.Textbox(label="Location")
    linkedin = gr.Textbox(label="LinkedIn")
    github = gr.Textbox(label="GitHub")
    role = gr.Textbox(label="Target Job Role")
    education = gr.Textbox(label="Education")
    programming = gr.Textbox(label="Programming Languages")
    libraries = gr.Textbox(label="Libraries & Frameworks")
    tools = gr.Textbox(label="Tools")
    databases = gr.Textbox(label="Databases")
    platforms = gr.Textbox(label="Platforms")
    project1 = gr.Textbox(label="Project 1")
    project2 = gr.Textbox(label="Project 2")
    project3 = gr.Textbox(label="Project 3")
    internship = gr.Textbox(label="Internship / Training")
    company = gr.Textbox(label="Company Name")
    position = gr.Textbox(label="Applying Position")
    
    btn = gr.Button("Generate")

    resume = gr.Textbox(label="Resume", lines=20)
    cover = gr.Textbox(label="Cover Letter", lines=20)

    resume_file = gr.File(label="Download Resume")
    cover_file = gr.File(label="Download Cover Letter")

    btn.click(
        generate,
        inputs=[
            name, phone, email, linkedin, github, location,
            role, education,
            programming, libraries, tools, databases, platforms,
            project1, project2, project3,
            internship, company, position
        ],
        outputs=[
            resume,
            cover,
            resume_file,
            cover_file
        ]
    )

resume_builder.launch()
```
