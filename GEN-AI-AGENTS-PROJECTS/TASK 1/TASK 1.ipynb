```python
import google.generativeai as genai
import time

# ==========================
# GEMINI API KEY
# ==========================

genai.configure(api_key="AQ.Ab8RN6KkiSjuWMORG7txBAhOu1ExCza6fZIvs6iuePdqp91aYw")

model = genai.GenerativeModel("gemini-2.5-flash")

print("=" * 60)
print("AI RESUME & COVER LETTER GENERATOR")
print("=" * 60)

# ==========================
# BASIC DETAILS
# ==========================

name = input("Enter Full Name: ")
phone = input("Enter Phone Number: ")
email = input("Enter Email: ")
linkedin = input("Enter LinkedIn URL: ")
github = input("Enter GitHub URL: ")
location = input("Enter Location (City, State): ")

# ==========================
# CAREER DETAILS
# ==========================

role = input("Enter Target Job Role: ")
education = input("Enter Education: ")

# ==========================
# SKILLS
# ==========================

programming = input("Programming Languages: ")
libraries = input("Libraries & Frameworks: ")
tools = input("Tools: ")
databases = input("Databases: ")
platforms = input("Platforms: ")

# ==========================
# PROJECTS
# ==========================

project1 = input("Project 1 Name: ")
project2 = input("Project 2 Name: ")
project3 = input("Project 3 Name: ")

# ==========================
# INTERNSHIP
# ==========================

internship = input("Internship / Training Name: ")

# ==========================
# COVER LETTER DETAILS
# ==========================

company_name = input("Company Name: ")
hiring_role = input("Applying For Position: ")

# ==========================
# RESUME PROMPT
# ==========================

resume_prompt = f"""
Create a professional ATS-friendly resume.

IMPORTANT RULES:

Use ONLY information provided.

DO NOT create fake projects.
DO NOT create fake internships.
DO NOT create fake achievements.
DO NOT create fake percentages.
DO NOT create fake skills.

Generate short professional summaries for projects and internship based on their names.

Candidate Details:

Name: {name}
Phone: {phone}
Email: {email}
LinkedIn: {linkedin}
GitHub: {github}
Location: {location}

Target Role:
{role}

Education:
{education}

Skills:

Programming Languages:
{programming}

Libraries & Frameworks:
{libraries}

Tools:
{tools}

Databases:
{databases}

Platforms:
{platforms}

Projects:

1. {project1}
2. {project2}
3. {project3}

Internship:
{internship}

Generate sections:

1. Header
2. Career Objective
3. Skills
4. Projects
   - Project Name
   - 2-3 line Summary
   - Technologies Used
5. Internship
   - Internship Name
   - 2-3 line Summary
6. Education

Keep resume professional.
"""

resume_response = model.generate_content(resume_prompt)

resume_text = resume_response.text

print("\n" + "=" * 60)
print("GENERATED RESUME")
print("=" * 60)

print(resume_text)

# ==========================
# SAVE RESUME
# ==========================

with open("Resume.txt", "w", encoding="utf-8") as file:
    file.write(resume_text)

print("\nResume saved as Resume.txt")

# ==========================
# WAIT FOR RATE LIMIT
# ==========================

print("\nGenerating Cover Letter...")
time.sleep(15)

# ==========================
# COVER LETTER PROMPT
# ==========================

cover_prompt = f"""
Write a professional cover letter.

Candidate Details:

Name: {name}
Phone: {phone}
Email: {email}
Location: {location}

Applying For:
{hiring_role}

Company:
{company_name}

Education:
{education}

Skills:

Programming Languages:
{programming}

Libraries & Frameworks:
{libraries}

Tools:
{tools}

Databases:
{databases}

Platforms:
{platforms}

Projects:
{project1}
{project2}
{project3}

Internship:
{internship}

Requirements:

1. Professional tone
2. Mention relevant skills
3. Mention projects briefly
4. Mention internship briefly
5. Explain interest in the role
6. Keep it concise
7. End professionally
"""

cover_response = model.generate_content(cover_prompt)

cover_text = cover_response.text

print("\n" + "=" * 60)
print("GENERATED COVER LETTER")
print("=" * 60)

print(cover_text)

# ==========================
# SAVE COVER LETTER
# ==========================

with open("CoverLetter.txt", "w", encoding="utf-8") as file:
    file.write(cover_text)

print("\nCover Letter saved as CoverLetter.txt")

print("\nProject Completed Successfully!")
```
