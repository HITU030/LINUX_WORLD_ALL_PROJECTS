# TASK 12 : AI Research Agent

```python
import ollama
import gradio as gr

MODEL = "llama3.2:latest"

def ollama_call(prompt):
    return ollama.chat(
        model=MODEL,
        messages=[{"role": "user", "content": prompt}]
    )["message"]["content"]


def research_agent(topic):

    prompt = f"""
Generate a research report on:

{topic}

Include:

1. Introduction
2. Key Concepts
3. Advantages
4. Challenges
5. Future Scope

Keep it simple, professional, and beginner-friendly.
"""

    result = ollama_call(prompt)

    with open("ResearchReport.txt", "w", encoding="utf-8") as f:
        f.write(result)

    return result, "ResearchReport.txt"


with gr.Blocks(title="AI Research Agent") as demo:

    gr.Markdown("# 🔍 AI Research Agent")

    topic = gr.Textbox(
        label="Research Topic",
        placeholder="Example: Artificial Intelligence in Healthcare",
        lines=2
    )

    btn = gr.Button("Generate Report")

    output = gr.Textbox(
        label="Research Report",
        lines=22
    )

    file = gr.File(label="Download Report")

    btn.click(
        research_agent,
        inputs=topic,
        outputs=[output, file]
    )

demo.launch()
```
