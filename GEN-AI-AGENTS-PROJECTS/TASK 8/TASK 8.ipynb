# TASK 8  : AI Code Assistant Tool

```python
import ollama
import gradio as gr

MODEL = "llama3.2:latest"

def ollama_call(prompt):
    return ollama.chat(
        model=MODEL,
        messages=[{"role": "user", "content": prompt}]
    )["message"]["content"]


def code_assistant(mode, content):

    prompt = f"""
Mode:
{mode}

User Content:
{content}

Instructions:

If Generate Code:
- Generate complete Python code.

If Explain Code:
- Explain line by line.

If Debug Code:
- Find errors.
- Provide corrected code.

If Optimize Code:
- Improve efficiency.
- Explain improvements.

If Generate Documentation:
- Generate professional documentation.

Keep the explanation beginner-friendly.
"""

    result = ollama_call(prompt)

    with open("CodeAssistantReport.txt", "w", encoding="utf-8") as f:
        f.write(result)

    return result, "CodeAssistantReport.txt"


with gr.Blocks(title="AI Code Assistant Tool") as demo:

    gr.Markdown("# 💻 AI Code Assistant Tool")

    mode = gr.Dropdown(
        [
            "Generate Code",
            "Explain Code",
            "Debug Code",
            "Optimize Code",
            "Generate Documentation"
        ],
        value="Generate Code",
        label="Select Mode"
    )

    content = gr.Textbox(
        label="Paste Code / Problem Statement",
        lines=12,
        placeholder="Example:\nWrite a Python program to reverse a string\n\nOR\nPaste your code here..."
    )

    btn = gr.Button("Generate")

    output = gr.Textbox(
        label="Result",
        lines=22
    )

    file = gr.File(label="Download Report")

    btn.click(
        code_assistant,
        inputs=[mode, content],
        outputs=[output, file]
    )

demo.launch()
```
