# TASK 3  : AI Study Assistant

```python
import ollama
import gradio as gr

MODEL = "llama3.2:latest"

def ollama_call(prompt):
    return ollama.chat(
        model=MODEL,
        messages=[{"role": "user", "content": prompt}]
    )["message"]["content"]


def study_assistant(topic, notes):

    prompt = f"""
You are an AI Study Assistant.

Topic:
{topic}

Notes:
{notes}

Generate:

1. Short Summary

2. Important Points

3. 5 Flashcards

4. 5 Multiple Choice Questions with Answers

5. Study Tips

Keep everything easy to understand.
"""

    result = ollama_call(prompt)

    with open("Generated_Notes.txt", "w", encoding="utf-8") as f:
        f.write(result)

    return result, "Generated_Notes.txt"


with gr.Blocks(title="AI Study Assistant") as demo:

    gr.Markdown("# 📚 AI Study Assistant")

    topic = gr.Textbox(label="Topic Name")

    notes = gr.Textbox(
        label="Paste Your Notes",
        lines=10,
        placeholder="Paste your study notes here..."
    )

    btn = gr.Button("Generate Study Material")

    output = gr.Textbox(
        label="Generated Study Material",
        lines=22
    )

    file = gr.File(label="Download Notes")

    btn.click(
        study_assistant,
        inputs=[topic, notes],
        outputs=[output, file]
    )

demo.launch()
```
