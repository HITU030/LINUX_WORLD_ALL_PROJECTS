# TASK 14 : AI Business Automation Agent

```python
import ollama
import gradio as gr

MODEL = "llama3.2:latest"

def ollama_call(prompt):
    return ollama.chat(
        model=MODEL,
        messages=[{"role": "user", "content": prompt}]
    )["message"]["content"]


def business_report(name, sales, expenses, customers):

    prompt = f"""
You are a business consultant.

Business Name:
{name}

Monthly Sales:
{sales}

Monthly Expenses:
{expenses}

Number of Customers:
{customers}

Generate:

1. Business Summary
2. Profit Analysis
3. Growth Opportunities
4. Cost Reduction Suggestions
5. Future Recommendations

Keep the response practical and professional.
"""

    result = ollama_call(prompt)

    with open("BusinessReport.txt", "w", encoding="utf-8") as f:
        f.write(result)

    return result, "BusinessReport.txt"


with gr.Blocks(title="AI Business Automation Agent") as demo:

    gr.Markdown("# 📊 AI Business Automation Agent")

    name = gr.Textbox(
        label="Business Name",
        placeholder="Example: Sharma Electronics"
    )

    sales = gr.Textbox(
        label="Monthly Sales (₹)",
        placeholder="Example: 500000"
    )

    expenses = gr.Textbox(
        label="Monthly Expenses (₹)",
        placeholder="Example: 320000"
    )

    customers = gr.Textbox(
        label="Number of Customers",
        placeholder="Example: 450"
    )

    btn = gr.Button("Generate Business Report")

    output = gr.Textbox(
        label="Business Analysis",
        lines=22
    )

    file = gr.File(label="Download Report")

    btn.click(
        business_report,
        inputs=[name, sales, expenses, customers],
        outputs=[output, file]
    )

demo.launch()
```
