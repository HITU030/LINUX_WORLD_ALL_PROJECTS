# TASK 13 : AI Job Search Agent

```python
import ollama
import gradio as gr

MODEL = "llama3.2:latest"

def ollama_call(prompt):
    return ollama.chat(
        model=MODEL,
        messages=[{"role": "user", "content": prompt}]
    )["message"]["content"]


def job_search_agent(education, skills, role):

    prompt = f"""
You are an AI Career Advisor.

Education:
{education}

Skills:
{skills}

Interested Role:
{role}

Generate:

1. Suitable Job Roles
2. Required Skills To Improve
3. Career Roadmap
4. Interview Preparation Tips

Keep the response practical and beginner-friendly.
"""

    result = ollama_call(prompt)

    with open("JobSearchReport.txt", "w", encoding="utf-8") as f:
        f.write(result)

    return result, "JobSearchReport.txt"


with gr.Blocks(title="AI Job Search Agent") as demo:

    gr.Markdown("# 💼 AI Job Search Agent")

    education = gr.Textbox(
        label="Education",
        placeholder="Example: B.Tech CSE"
    )

    skills = gr.Textbox(
        label="Skills",
        lines=3,
        placeholder="Example: Python, SQL, Machine Learning, Git"
    )

    role = gr.Textbox(
        label="Interested Role",
        placeholder="Example: AI Engineer"
    )

    btn = gr.Button("Generate Career Report")

    output = gr.Textbox(
        label="Career Guidance",
        lines=22
    )

    file = gr.File(label="Download Report")

    btn.click(
        job_search_agent,
        inputs=[education, skills, role],
        outputs=[output, file]
    )

demo.launch()
```
