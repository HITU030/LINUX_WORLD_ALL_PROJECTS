```python
import google.generativeai as genai

genai.configure(api_key="AQ.Ab8RN6KkiSjuWMORG7txBAhOu1ExCza6fZIvs6iuePdqp91aYw")

model = genai.GenerativeModel("gemini-2.5-flash")

print("=" * 50)
print("AI JOB SEARCH AGENT")
print("=" * 50)

education = input("Enter Education: ")
skills = input("Enter Skills: ")
interested_role = input("Interested Role: ")

prompt = f"""
You are an AI Career Advisor.

Education:
{education}

Skills:
{skills}

Interested Role:
{interested_role}

Generate:

1. Suitable Job Roles

2. Required Skills To Improve

3. Career Roadmap

4. Interview Preparation Tips

Keep response practical and beginner friendly.
"""

response = model.generate_content(prompt)

result = response.text

print("\n")
print(result)

with open("JobSearchReport.txt", "w", encoding="utf-8") as file:
    file.write(result)

print("\nReport saved as JobSearchReport.txt")
```
