# TASK 4  : Prompt-Based Coding Assistant


```python
import ollama
import gradio as gr

MODEL = "llama3.2:latest"

def ollama_call(prompt):
    return ollama.chat(
        model=MODEL,
        messages=[{"role": "user", "content": prompt}]
    )["message"]["content"]


def coding_assistant(problem):

    prompt = f"""
You are an expert coding assistant.

Problem:
{problem}

Generate:

1. Python Code

2. Code Explanation

3. Input Output Example

4. Improvement Suggestions

Keep the explanation beginner-friendly.
"""

    result = ollama_call(prompt)

    with open("CodingAssistant.txt", "w", encoding="utf-8") as f:
        f.write(result)

    return result, "CodingAssistant.txt"


with gr.Blocks(title="AI Coding Assistant") as demo:

    gr.Markdown("# 💻 AI Coding Assistant")

    problem = gr.Textbox(
        label="Coding Problem",
        lines=6,
        placeholder="Describe your coding problem..."
    )

    btn = gr.Button("Generate Solution")

    output = gr.Textbox(
        label="Generated Solution",
        lines=22
    )

    file = gr.File(label="Download Solution")

    btn.click(
        coding_assistant,
        inputs=problem,
        outputs=[output, file]
    )

demo.launch()
```
