# TASK 5  : AI Email Automation System

```python
import ollama
import gradio as gr

MODEL = "llama3.2:latest"

def ollama_call(prompt):
    return ollama.chat(
        model=MODEL,
        messages=[{"role": "user", "content": prompt}]
    )["message"]["content"]


def generate_email(email_type, name, recipient, company, purpose):

    prompt = f"""
Generate a professional email.

Email Type:
{email_type}

Sender:
{name}

Recipient:
{recipient}

Company:
{company}

Purpose:
{purpose}

Generate:

1. Subject Line
2. Email Body
3. Professional Closing
4. Signature

Keep it professional.
"""

    result = ollama_call(prompt)

    with open("GeneratedEmail.txt", "w", encoding="utf-8") as f:
        f.write(result)

    return result, "GeneratedEmail.txt"


with gr.Blocks(title="AI Email Automation System") as demo:

    gr.Markdown("# 📧 AI Email Automation System")

    email_type = gr.Dropdown(
        [
            "HR Email",
            "Job Application",
            "Leave Request",
            "Sales Email",
            "Customer Support Email",
            "Custom Email"
        ],
        label="Email Type",
        value="HR Email"
    )

    name = gr.Textbox(label="Your Name")
    recipient = gr.Textbox(label="Recipient Name")
    company = gr.Textbox(label="Company Name")

    purpose = gr.Textbox(
        label="Purpose of Email",
        lines=5
    )

    btn = gr.Button("Generate Email")

    output = gr.Textbox(
        label="Generated Email",
        lines=20
    )

    file = gr.File(label="Download Email")

    btn.click(
        generate_email,
        inputs=[email_type, name, recipient, company, purpose],
        outputs=[output, file]
    )

demo.launch()
```
