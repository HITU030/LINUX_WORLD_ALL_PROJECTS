# TASK 2  : Business Idea Generator AI

```python
import ollama
import gradio as gr

MODEL = "llama3.2:latest"

def ollama_call(prompt):
    return ollama.chat(
        model=MODEL,
        messages=[{"role": "user", "content": prompt}]
    )["message"]["content"]


def generate_idea(domain, budget, country, audience):

    prompt = f"""
Generate a startup business idea.

Business Name:
{domain}

Budget:
{budget}

Target Country:
{country}

Target Audience:
{audience}

Generate:

1. Startup Name
2. Problem Statement
3. Solution
4. Target Audience
5. Revenue Model
6. Market Validation
7. Future Scope
8. Why This Idea Can Work

Keep the response professional.
"""

    idea = ollama_call(prompt)

    with open("BusinessIdea.txt", "w", encoding="utf-8") as f:
        f.write(idea)

    return idea, "BusinessIdea.txt"


with gr.Blocks(title="AI Business Idea Generator") as demo:

    gr.Markdown("# 💡 AI Business Idea Generator")

    business = gr.Textbox(label="Business Name / Domain")

    with gr.Row():
        budget = gr.Dropdown(
            ["Low", "Medium", "High"],
            label="Budget",
            value="Medium"
        )

        country = gr.Textbox(label="Target Country")

    audience = gr.Textbox(label="Target Audience")

    btn = gr.Button("Generate Idea")

    output = gr.Textbox(
        label="Generated Business Idea",
        lines=20
    )

    file = gr.File(label="Download Business Idea")

    btn.click(
        generate_idea,
        inputs=[business, budget, country, audience],
        outputs=[output, file]
    )

demo.launch()
```
