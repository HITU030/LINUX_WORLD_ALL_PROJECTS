# TASK 10 : AI Image Prompt Generator

```python
import ollama
import gradio as gr

MODEL = "llama3.2:latest"

def ollama_call(prompt):
    return ollama.chat(
        model=MODEL,
        messages=[{"role": "user", "content": prompt}]
    )["message"]["content"]


def generate_prompt(idea):

    prompt = f"""
Create one professional AI image generation prompt.

Image Idea:
{idea}

Make it detailed, realistic, cinematic and ready to use in
Midjourney, DALL·E or Stable Diffusion.
"""

    result = ollama_call(prompt)

    with open("ImagePrompt.txt", "w", encoding="utf-8") as f:
        f.write(result)

    return result, "ImagePrompt.txt"


with gr.Blocks(title="AI Image Prompt Generator") as demo:

    gr.Markdown("# 🎨 AI Image Prompt Generator")

    idea = gr.Textbox(
        label="Image Idea",
        lines=3,
        placeholder="Example: A futuristic cyberpunk city at night with flying cars"
    )

    btn = gr.Button("Generate Prompt")

    output = gr.Textbox(
        label="Generated Prompt",
        lines=15
    )

    file = gr.File(label="Download Prompt")

    btn.click(
        generate_prompt,
        inputs=idea,
        outputs=[output, file]
    )

demo.launch()
```
