# TASK 6  : AI Content Creation Platform

```python
import ollama
import gradio as gr

MODEL = "llama3.2:latest"

def ollama_call(prompt):
    return ollama.chat(
        model=MODEL,
        messages=[{"role": "user", "content": prompt}]
    )["message"]["content"]


def generate_content(topic, audience, tone):

    prompt = f"""
You are a professional content creator.

Topic:
{topic}

Target Audience:
{audience}

Tone:
{tone}

Generate:

1. Blog Article (300-500 words)
2. LinkedIn Post
3. Instagram Caption
4. 10 Relevant Hashtags
5. SEO Keywords

Keep the content engaging and professional.
"""

    result = ollama_call(prompt)

    with open("ContentCreation.txt", "w", encoding="utf-8") as f:
        f.write(result)

    return result, "ContentCreation.txt"


with gr.Blocks(title="AI Content Creation Platform") as demo:

    gr.Markdown("# ✍️ AI Content Creation Platform")

    topic = gr.Textbox(label="Topic")
    audience = gr.Textbox(label="Target Audience")

    tone = gr.Dropdown(
        ["Professional", "Friendly", "Marketing"],
        label="Tone",
        value="Professional"
    )

    btn = gr.Button("Generate Content")

    output = gr.Textbox(
        label="Generated Content",
        lines=22
    )

    file = gr.File(label="Download Content")

    btn.click(
        generate_content,
        inputs=[topic, audience, tone],
        outputs=[output, file]
    )

demo.launch()
```
