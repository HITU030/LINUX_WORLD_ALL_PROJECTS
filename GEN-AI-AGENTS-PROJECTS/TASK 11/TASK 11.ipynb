# TASK 11 : Personal AI Assistant

```python
import ollama
import gradio as gr

MODEL = "llama3.2:latest"

def ollama_call(prompt):
    return ollama.chat(
        model=MODEL,
        messages=[{"role": "user", "content": prompt}]
    )["message"]["content"]


def personal_assistant(tasks):

    prompt = f"""
You are a personal productivity assistant.

Tasks:
{tasks}

Generate:

1. Priority Order
2. Daily Plan
3. Productivity Tips

Keep the response short, practical, and beginner-friendly.
"""

    result = ollama_call(prompt)

    with open("PersonalAssistant.txt", "w", encoding="utf-8") as f:
        f.write(result)

    return result, "PersonalAssistant.txt"


with gr.Blocks(title="Personal AI Assistant") as demo:

    gr.Markdown("# 🧠 Personal AI Assistant")

    tasks = gr.Textbox(
        label="Enter Your Tasks",
        lines=6,
        placeholder="""Example:
- Complete Python assignment
- Prepare for interview
- Go to gym
- Read AI notes
- Finish project"""
    )

    btn = gr.Button("Generate Plan")

    output = gr.Textbox(
        label="AI Productivity Plan",
        lines=18
    )

    file = gr.File(label="Download Plan")

    btn.click(
        personal_assistant,
        inputs=tasks,
        outputs=[output, file]
    )

demo.launch()
```
