# TASK 7  : AI Chatbot for Business

```python
import ollama
import gradio as gr

MODEL = "llama3.2:latest"

def ollama_call(prompt):
    return ollama.chat(
        model=MODEL,
        messages=[{"role": "user", "content": prompt}]
    )["message"]["content"]


chat_history = ""


def chatbot(business_name, business_type, timing, location, services, question):
    global chat_history

    prompt = f"""
You are a professional business customer support chatbot.

Business Name:
{business_name}

Business Type:
{business_type}

Business Timing:
{timing}

Location:
{location}

Services:
{services}

Previous Conversation:
{chat_history}

Customer Question:
{question}

Give a short, professional and helpful reply.
"""

    answer = ollama_call(prompt)

    chat_history += f"\nCustomer: {question}\nBot: {answer}\n"

    with open("ChatHistory.txt", "w", encoding="utf-8") as f:
        f.write(chat_history)

    return answer, "ChatHistory.txt"


with gr.Blocks(title="AI Business Chatbot") as demo:

    gr.Markdown("# 🤖 AI Business Chatbot")

    business_name = gr.Textbox(
        label="Business Name",
        placeholder="Example: Sharma Electronics"
    )

    business_type = gr.Textbox(
        label="Business Type",
        placeholder="Example: Mobile Repair Shop"
    )

    timing = gr.Textbox(
        label="Business Timing",
        placeholder="Example: 10 AM - 8 PM"
    )

    location = gr.Textbox(
        label="Location",
        placeholder="Example: Jaipur, Rajasthan"
    )

    services = gr.Textbox(
        label="Services Offered",
        lines=3,
        placeholder="Example: Mobile Repair, Screen Replacement, Accessories, Software Update"
    )

    question = gr.Textbox(
        label="Customer Question",
        lines=2,
        placeholder="Example: What are your shop timings?"
    )

    btn = gr.Button("Ask")

    answer = gr.Textbox(
        label="Bot Reply",
        lines=8
    )

    file = gr.File(label="Download Chat History")

    btn.click(
        chatbot,
        inputs=[
            business_name,
            business_type,
            timing,
            location,
            services,
            question
        ],
        outputs=[answer, file]
    )

demo.launch()
```
