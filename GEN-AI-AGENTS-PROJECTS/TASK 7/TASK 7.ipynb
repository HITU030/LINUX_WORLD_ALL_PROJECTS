```python
import google.generativeai as genai

# ==========================
# GEMINI API KEY
# ==========================

genai.configure(api_key="AQ.Ab8RN6KkiSjuWMORG7txBAhOu1ExCza6fZIvs6iuePdqp91aYw")

model = genai.GenerativeModel("gemini-2.5-flash")

print("=" * 60)
print("AI BUSINESS CHATBOT")
print("=" * 60)

# ==========================
# BUSINESS DETAILS
# ==========================

business_name = input("Business Name: ")
business_type = input("Business Type: ")
timing = input("Business Timing: ")
location = input("Location: ")
services = input("Services Offered: ")

print("\nType 'exit' to stop chatbot.\n")

chat_history = ""

while True:

    question = input("Customer: ")

    if question.lower() == "exit":
        break

    prompt = f"""
    You are a customer support chatbot.

    Business Name:
    {business_name}

    Business Type:
    {business_type}

    Timing:
    {timing}

    Location:
    {location}

    Services:
    {services}

    Previous Conversation:
    {chat_history}

    Customer Question:
    {question}

    Give a professional and short reply.
    """

    response = model.generate_content(prompt)

    answer = response.text

    print("\nBot:", answer)

    chat_history += f"\nCustomer: {question}\nBot: {answer}\n"

# ==========================
# SAVE CHAT
# ==========================

with open("ChatHistory.txt", "w", encoding="utf-8") as file:
    file.write(chat_history)

print("\nChat saved as ChatHistory.txt")
```
