# TASK 1: - COMPLETE MENU FOR pywhatkit api

```python
import pywhatkit

def google_search():
    search = input("enter what you want to search")
    pywhatkit.search(search)
    print("searching.............")
    
def youtube_video_play():
    search = input("enter youtube video title :")
    pywhatkit.playonyt(search)
    print("redirecting to the youtube......")

def wikipedia_summary():
    topic_name = input("enter topic name ")
    summary_lines = int(input("enter how many lines summary you want"))
    summary = pywhatkit.info(topic_name , summary_lines)
    print(summary)
    
def whatsapp_msg_sender():
    contact_number = "+91" + input("enter number who you want to send msg")
    message = input("enter your message: ")
    pywhatkit.sendwhatmsg_instantly(
    contact_number,
    message
    )
    print("message bhej diya gya h ")

while True:
    
    print("""
    ==== ALL USES OF PYWHATAPI ====

    == CHOOSE ONE OPTION ==
    
    1. Google search
    2. Youtube video player
    3. A topic summary
    4. whatsapp message sender
    5. End / Exit

    """
         )

    choice = int(input("enter your choice :- "))

    if choice == 1:
        google_search()
        
    elif choice == 2:
        youtube_video_play()
        
    elif choice == 3:
        wikipedia_summary()
        
    elif choice == 4:
        whatsapp_msg_sender()
        
    elif choice == 5:
        print("THANKS FOR USING THIS MENU !!!")
        break
        
    else:
        print("sir you select wrong choice")
```

```python
#for login in account login  
from instagrapi import Client

def login_and_save_session():
    
    username = input("Enter Username : ")
    password = input("Enter Password : ")
    print("logining into your account......")
    cl = Client()

    cl.login(username, password)
    

    cl.dump_settings("session.json")

    print("Session Saved Successfully")
```

```python
#sesssion loaddingg
from instagrapi import Client

def load_session():

    cl = Client()

    cl.load_settings("session.json")

    return cl
```

```python
# acc. info
def account_info():

    cl = load_session()

    print(cl.account_info())
```

```python
# photuuu upload
def upload_photo():

    cl = load_session()

    image_path = input("Enter Image Path : ")

    caption = input("Enter Caption : ")

    media = cl.photo_upload(
        image_path,
        caption
    )

    print(media)

    print("Photo Uploaded Successfully")
```

```python
# video uploadd ke liye
def upload_video():

    cl = load_session()
    
    video_path = input("Enter Video Path : ")

    caption = input("Enter Caption : ")

    media = cl.video_upload(
        video_path,
        caption
    )

    print(media)

    print("Video Uploaded Successfully")
```

```python
# storyyy ke liye
def upload_image_story():

    cl = load_session()

    image_path = input("Enter Story Path : ")

    cl.photo_upload_to_story(
        image_path
    )

    print("Story Uploaded Successfully")
```

```python
# story pe video 
def upload_video_on_story():

    cl = load_session()
    
    video_path = input("Enter Story Video Path : ")

    cl.video_upload_to_story(
        video_path
    )

    print("Story Video Uploaded Successfully")
```

```python
# msg bhejne ke liyeeeee
def send_dm():

    cl = load_session()

    target_username = input("Enter Username : ")

    message = input("Enter Message : ")

    user_id = cl.user_id_from_username(
        target_username
    )

    cl.direct_send(
        message,
        [user_id]
    )

    print("Message Sent")
```

```python
#follow 
def follow_user():

    cl = load_session()

    target_username = input("Enter Username : ")

    user_id = cl.user_id_from_username(
        target_username
    )

    cl.user_follow(user_id)

    print("User Followed")
```

```python
#unfollow a ussr
def unfollow_user():

    cl = load_session()
    
    target_username = input("Enter Username : ")

    user_id = cl.user_id_from_username(
        target_username
    )

    cl.user_unfollow(user_id)

    print("User Unfollowed")
```

```python
# follower list kud ki
def followers_list():

    cl = load_session()

    followers = cl.user_followers(
        cl.user_id
    )

    print(followers)
```

```python
#following list khud ki
def following_list():

    cl = load_session()
    
    following = cl.user_following(
        cl.user_id
    )

    print(following)
```

```python
# for search user
def search_user():

    cl = load_session()
    
    target = input("Search User : ")

    user = cl.user_info_by_username(
        target
    )

    print(user)
```

```python
#dusre ke follower
def get_others_followers():

    cl = load_session()

    username = input("Enter Username : ")

    user_id = cl.user_id_from_username(
        username
    )

    followers = cl.user_followers(
        user_id
    )

    print(followers)
```

```python
# following dusre ki
def get_others_following():

    cl = load_session()

    username = input("Enter Username : ")

    user_id = cl.user_id_from_username(
        username
    )

    following = cl.user_following(
        user_id
    )

    print(following)
```

```python
#downloading another person post
def download_by_url():

    cl = load_session()

    url = input("Enter Post URL : ")

    media_pk = cl.media_pk_from_url(
        url
    )

    cl.photo_download(
        media_pk,
        "downloads"
    )

    print("Downloaded")
```
