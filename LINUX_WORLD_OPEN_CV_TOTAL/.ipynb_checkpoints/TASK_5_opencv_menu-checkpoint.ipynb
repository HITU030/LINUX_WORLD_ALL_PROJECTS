```python

# operation on image
def operation_on_image():
    import cv2
    import matplotlib.pyplot as plt
    
    def resized_image():
        
        uploaded_image = input("upload your image path")
        print("enter croping cordinates")
        a = int(input("Enter a coordinate: "))
        b = int(input("Enter b coordinate: "))
        image_1 = cv2.imread(uploaded_image)
        image_1 = cv2.cvtColor(image_1, cv2.COLOR_BGR2RGB)
        image_2 = cv2.resize(image_1, (a,b))
        
        fig, axes = plt.subplots(1, 2)
        axes[0].imshow(image_1)
        axes[0].axis("off")
        axes[1].imshow(image_2)
        axes[1].axis("off")
        plt.show()
    
    def croping_image():
        
        uploaded_image = input("upload your image path")
        image = cv2.imread(uploaded_image)
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        
        height, width = image_rgb.shape[:2]
        print("given image size")
        print(f"Image Width : {width}")
        print(f"Image Height: {height}")
    
        print("enter cordinates for croping img")
        start_row = int(input("Enter start row-"))
        end_row = int(input("Enter end row-- "))
        start_col = int(input("Enter start column -"))
        end_col = int(input("Enter end column-  "))
        cropped = image_rgb[start_row:end_row, start_col:end_col]
        
        fig , axes = plt.subplots(1,2 , figsize = (10,4))
        axes[0].imshow(image_rgb)
        axes[0].set_title("original image")
        axes[0].axis("off")
        axes[1].imshow(cropped)
        axes[1].set_title("cropped image")
        axes[1].axis("off")
    
    def rotate_image():
        uploaded_image = input("upload your image path")
        image = cv2.imread(uploaded_image)
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        
        height , width = image.shape[:2]
        center = (width//2 , height//2)
        angle = int(input("enter the angle you want to rotate"))
        
        rotation_matrix = cv2.getRotationMatrix2D(center,angle,1)
        rotated = cv2.warpAffine(image, rotation_matrix , (width,height))
        rotated_rgb = cv2.cvtColor(rotated , cv2.COLOR_BGR2RGB)
        
        fig , axes = plt.subplots(1,2 , figsize = (10,4))
        axes[0].imshow(image_rgb)
        axes[0].axis("off")
        axes[1].imshow(rotated_rgb)
        axes[1].axis("off")
        plt.show()
    
    def gery_scale():
        uploaded_image = input("upload your image path")
        image = cv2.imread(uploaded_image) 
        grey = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
        hsv = cv2.cvtColor(image , cv2.COLOR_BGR2HSV)
        binary = cv2.threshold(grey , 127 , 255 , cv2.THRESH_BINARY ) 
        image_rgb = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
        
        fig , axes = plt.subplots(1,4 , figsize = (18,4))
        axes[0].imshow(image_rgb)
        axes[0].axis("off")
        axes[1].imshow(grey , cmap = "grey")
        axes[1].axis("off")
        plt.show()
    
    while True:
        
        print("""
        
        -----choose one operation you want to perform on your uploaded image -----
        
         1. resize image
         2. crop image
         3. rotate image
         4. gery image
         5. exittttt
         """
             )
        
        choice = int(input("enter your choice"))
        
        if choice == 1:
            resized_image()
            
        elif choice == 2:
            croping_image()
            
        elif choice == 3:
            rotate_image()
            
        elif choice == 4:
            gery_scale()
            
        elif choice == 5:
            print(":::: tata byy byy :::::")
            break
            
        else:
            print("you are wrongg number")





# operation on vidoe 

def operation_on_video():
    import cv2
    face_detector = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
    
    def rectangle_around_face():
        print("""
        camera opening......
        .
        """)
        
        cap = cv2.VideoCapture(0)
        print("camera is running......")
        while True:
            succes , frame = cap.read()
            if not succes:
                break
            grey = cv2.cvtColor(frame , cv2.COLOR_BGR2RGB)  
            face = face_detector.detectMultiScale(grey , 1.1 ,5)
            for (x, y, w, h) in face:
                cv2.rectangle(frame , (x,y) , (x+w , y+w) , (0,255,0) , 2)
                #cv2.circle(frame ,(x,y) , 50 , (255,0,0) , 2)
                cv2.imshow("face detection" , frame)
                
            if cv2.waitKey(1) == ord('q'):
                break
                    
        cap.release()
        cv2.destroyAllWindows()
    
    
    
    
    def photo_on_face():
        
        uploaded_emoji = input("enter file location")
        emoji = cv2.imread(uploaded_emoji)
        cap = cv2.VideoCapture(0)
        
        while True:
            succes , frame = cap.read()
            if not succes:
                break
            grey = cv2.cvtColor(frame , cv2.COLOR_BGR2RGB)  
            face = face_detector.detectMultiScale(grey , 1.1 ,5)
            for (x, y, w, h) in face:
                emoji_resize = cv2.resize(emoji , (w,h))
                frame [y: y+h , x:x+w] =emoji_resize 
                cv2.imshow("emoji face" , frame)
        
            if cv2.waitKey(1) == ord('q'):
                break
                    
        cap.release()
        cv2.destroyAllWindows()
    
    
    
    
    def tect_on_video():
        cap = cv2.VideoCapture(0)
        while True:
            succes , frame = cap.read()
            if not succes:
                break
            cv2.putText(frame , "my face" , (20,40) , cv2.FONT_HERSHEY_DUPLEX , 1 , (255,0,0) ,2)
            cv2.putText(frame , "helloo" , (50,70) , cv2.FONT_HERSHEY_SIMPLEX , 1 , (255,0,0) ,2)
            cv2.imshow("TEXT" , frame)
            if cv2.waitKey(1) == ord('q'):
                break
                
        cap.release()
        cv2.destroyAllWindows()
        
    
    
    
    def filter_on_face():
        
        cap = cv2.VideoCapture(0)
        mode = "color"
        
        while True:
            succes , frame = cap.read()
            if not succes:
                break
                
            if mode == "grey":
                frame = cv2.cvtColor(frame , cv2.COLOR_BGR2GRAY)
                
            elif mode == "red":
                frame[:,:,0] = 0
                frame[:,:,1] = 0
            cv2.imshow("webcam filter", frame)
            key = cv2.waitKey(1)
                
            if key == ord("g"):
                mode = "grey"
                
            elif key == ord("r"):
                mode = "red"   
                
            elif key == ord("c"):
                mode = "color"
                
            elif key == ord("q"):
                break
                    
        cap.release()
        cv2.destroyAllWindows()
    
    
    
    def face_count():
        
        cap = cv2.VideoCapture(0)
        
        while True:
            succes , frame = cap.read()
            if not succes:
                break
            gray = cv2.cvtColor(frame , cv2.COLOR_BGR2GRAY)
            face = face_detector.detectMultiScale(
                gray,
                scaleFactor=1.1,
                minNeighbors=5, 
                minSize=(30, 30),        
            )
            for (x, y, w, h) in face:
                cv2.rectangle(
                    frame ,
                    (x,y) ,
                    (x+w,y+h) , 
                    (0,255,0),
                    2
                )
            count_text = f"faces : {len(face)}"
            cv2.rectangle(
                frame ,
                (0,0) ,
                (200 , 50) ,
                (0,255,0) ,
                -1
                            )
            cv2.putText(
                frame,
                count_text,
                (10,35),
                cv2.FONT_HERSHEY_SIMPLEX,
                1,
                (255,0,0),
                2
            )
            cv2.imshow("FACE COUNTER", frame)   
            if cv2.waitKey(1) & 0xff == ord('q'):
                break 
            print("final face count" , len(face))
            
        cap.release()
        cv2.destroyAllWindows()  

    while True:
        
        print("""
        
        -----choose one -----
        
         1. rectangle on face
         2. photo on face
         3. text on video top-side
         4. filters on video
         5. Face counting
         6. exittttt
         """
             )
        
        choice = int(input("enter your choice"))
        
        if choice == 1:
            rectangle_around_face()
            
        elif choice == 2:
            photo_on_face()
            
        elif choice == 3:
            tect_on_video()
            
        elif choice == 4:
            filter_on_face()
        
        elif choice == 5:
            face_count()
            
        elif choice == 6:
            print(":::: tata byy byy :::::")
            break
            
        else:
            print("you are wrongg number")

        
while True:
    print("""
    
    +=+=+ select one +=+=+
    
    1. IMAGE EDITING
    2. VIDEO EDITING
    3. exxit
    """)
    
    choice = int(input("enter your choice"))
            
    if choice == 1:
        operation_on_image()
        
    elif choice == 2:
        operation_on_video()
        
    elif choice == 3:
        print(":::: tata byy byy :::::")
        break
        
    else:
        print("you are wrongg number")
```
