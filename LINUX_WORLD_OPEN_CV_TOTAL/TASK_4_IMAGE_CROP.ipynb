```python
import cv2
import matplotlib.pyplot as plt
uploaded_image = input("upload your image path")
image = cv2.imread(uploaded_image)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

height, width = image_rgb.shape[:2]
print("given image size")
print(f"Image : {width} X {height}")

print("\nCrop Range Available")

print(f"Start Row : 0 to {height-1}")
print(f"End Row   : 1 to {height}")

print(f"Start Col : 0 to {width-1}")
print(f"End Col   : 1 to {width}")

print("enter cordinates for croping img")
start_row = int(input("Enter start row-"))
end_row = int(input("Enter end row-- "))
start_col = int(input("Enter start column -"))
end_col = int(input("Enter end column-  "))
cropped = image_rgb[start_row:end_row, start_col:end_col]

fig , axes = plt.subplots(1,2 , figsize = (10,4))
axes[0].imshow(image_rgb)
axes[0].set_title("original image")
axes[0].axis("off")
axes[1].imshow(cropped)
axes[1].set_title("cropped image")
axes[1].axis("off")
```
