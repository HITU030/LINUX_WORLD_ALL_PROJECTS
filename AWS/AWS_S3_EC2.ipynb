```python
def list():
    list_buckets = s3.list_buckets()
    list_buckets
    for i in list_buckets['Buckets']:
     print(i['Name'])
```

```python
def upload():
    file_path = input("Enter the file's path : ")
    bucket_name = input("Enter the bucket name : ")
    object_name = input("Enter the object name : ")
    s3.upload_file(file_path, bucket_name, object_name)
    print("File uploaded successfully")
```

```python
def download():
    bucket = input("Enter the bucket name : ")
    object_name = input("Enter the object name : ")
    local_file = input("save as : ")
    s3.download_file(bucket, object_name, local_file)
    print("File downloaded successfully")
```

```python
def delete():
    bucket = input("Enter bucket name: ")
    file_name = input("Enter file name to delete: ")

    s3.delete_object(
    Bucket=bucket,
    Key=file_name
    )

    print("File deleted successfully")
```
