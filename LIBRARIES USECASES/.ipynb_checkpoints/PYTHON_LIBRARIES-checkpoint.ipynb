# LIBRARIES

```python
import pyttsx3
def text_to_speech():
    engine = pyttsx3.init()
    speak = input("enter what you want to speak -: ")
    engine.say(speak)

    engine.runAndWait()


import speech_recognition as sr
def speech_to_text():
    recog = sr.Recognizer()

    with sr.Microphone() as source:
        print("Speak Now...")
        audio = recog.listen(source)

    text = recog.recognize_google(audio)
    print("You Said:", text)


import pyjokes
def random_joke():
    joke = pyjokes.get_joke()
    print("\n", joke)


import qrcode
def qr_code_generator():
    img = qrcode.make("https://www.google.com/")

    img.save("google_search.png")


import random
def password_generator():
    chars = "abcdefghijklmnopqrstuvwxyz123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"

    password = ""

    for i in range(8):
        password += random.choice(chars)

    print("Password:", password)
    
```

```python
while True:

    print("""
    ===== menu for python librarriess=====

    1. text to Speech
    2. Speech to Text
    3. QR Code Generator
    4. Random Joke
    5. Password Generator
    6. EXIT.

    """)

    choice = int(input("Enter Your Choice"))

    if choice == 1:
        text_to_speech()

    elif choice == 2:
        speech_to_text()

    elif choice == 3:
        qr_code_generator()

    elif choice == 4:
        random_joke()

    elif choice == 5:
        password_generator()
        
    elif choice == 6:
        print("BYE!!!")
        break

    else:
        print("Invalid Choice")
   
```
